Importing the libararies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data collection and data Processing

```python
#loading tha data frame from csv file to pandas dataframe
car_dataset=pd.read_csv('/content/car data.csv')
car_dataset.head()
```

```python
#checking the number of rows and columns
car_dataset.shape
```

```python
#getting some information about the dataset
car_dataset.info()
```

```python
#checking the number of missing values
car_dataset.isnull().sum()
```

```python
#checking the didtribution of categorial data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the categorial data

```python
#encoding fuel_type column
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2},'Seller_Type':{'Dealer':0,'Individual':1},'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_dataset.head()
```

Spilitting the data into training data and the test data

```python
X=car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y=car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

Spilitting Training and test data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.1,random_state=2)
```

Model Training

1. Linear Regesion

```python
#loading the linear regression model
lin_reg_model=LinearRegression()
lin_reg_model.fit(X_train,Y_train)
```

Model Evalution

```python
#prediction on training moddel
training_data_Prediction=lin_reg_model.predict(X_train)
```

```python
#R square error
error_score= metrics.r2_score(Y_train,training_data_Prediction)
print('R square error:', error_score)
```

Visualizing the actual price and predict price

```python
plt.scatter(Y_train,training_data_Prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs Predicted price")
plt.show()
```

Model Evalution

```python
#prediction on test moddel
test_data_Prediction=lin_reg_model.predict(X_test)
```

```python
#R square error
error_score= metrics.r2_score(Y_test,test_data_Prediction)
print('R square error:', error_score)
```

Visualizing the actual price and predict price

```python
plt.scatter(Y_test,test_data_Prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs Predicted price")
plt.show()
```

Model Training

Model Evalution
