Importing the libararies

```python
#importing the image module from matplot library
import matplotlib.image as mping
import matplotlib.pyplot as plt
```

Reading the image

```python
#loading an image through matplotlib. image module
img=mping.imread('/content/maxresdefault.jpg')
print(img)
```

```python
type(img)
print(img.shape)
```

Displaying the image

```python
#displaying the image from numpy array
img_plot=plt.imshow(img)
plt.show()
```

Resizing the image and saving it

```python

#image sholud be in same size
from PIL import Image
img=Image.open('/content/maxresdefault.jpg')
img_resized=img.resize((200,200))
img_resized.save('cat_image_resize.jpg')
```

Displayig the resizing image

```python
#displaying the image from numpy array after resizing it
img_res=mping.imread('/content/cat_image_resize.jpg') #reading 
img_res_plot=plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting the RGB images to Grayscale image using open CV

```python
#Importing open cv lib
import cv2
img=cv2.imread('/content/maxresdefault.jpg')
type(img)
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #IT SHOULD BE done in numpy array not in file
```

```python
type(grayscale_image)
grayscale_image.shape #converting to gray scale since '3' is not come
```

```python
from google.colab.patches import cv2_imshow
```

```python
#displaying the image
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale imange
cv2.imwrite('cat_grayscale_image.jpg',grayscale_image)
```
