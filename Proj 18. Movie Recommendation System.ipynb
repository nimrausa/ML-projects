Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib#close matches   name of favourite movie by user , user give diffrent spelling name so find the similarity due to apelling incorect
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity # help to find out similarity scor
```

Data Collection and pre preocessing

```python
#loading the csv file to pandas data frame
movies_data= pd.read_csv('/content/movies.csv')

```

```python
# printomg the five rows of data frame
movies_data.head()
```

```python
movies_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features=['genres','keywords','original_language','tagline','cast','director','popularity']
print(selected_features)
```

```python
#replacing the null values with null string
for features in selected_features:
  movies_data[features]=movies_data[features].fillna(' ')#filling with space value
```

```python
#combing all the 5 selected features
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['original_language']+' '
+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+str(movies_data ['popularity'])
combined_features.shape
print(combined_features)
```

```python
#convert the text data into the numerical data


vectorizer= TfidfVectorizer()
feature_vectors= vectorizer.fit_transform(combined_features) 
print(feature_vectors)
type(feature_vectors)
feature_vectors.shape

```

Cosine similarity

```python
#getting the similarity score using cosine similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
print(similarity.shape)
#for each movie we will compare wih all the other movies. For eg. Batman --
#it will take Batman movie and it will try to compare with the other movies there in the data set
#it will give similarity score of each movie
```

Movie name from user

```python
#getting the movie name from the user
movie_name=input('Enter the movie name :')

#creating the list with all the movie names given in the datset
list_of_all_titles= movies_data['title'].tolist()# it will compare movie name'Batman' with the list of movies
print(movie_name)
```

```python
#finding the close match for the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
close_match=find_close_match[0] #chossing the first value For eg: Batmab dark night, Batman dark night begins, Batman -----choose Batman
print(close_match)
```

```python
#finding the index of the movie with the title
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0] #locate the name of the movie and its corresponding index at first[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie])) # to run the loop in a particularly list
#if the movie is same the similarity_score is high and vice-versa
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on thier similarity score
sorted_similarity_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)# to sort or arrange the elements of the list
#x:x[1]--particular list similarity score, reverse: want to arrange value of the similarity score for highest to lowest
print(sorted_similarity_movies)
```

```python
#print the name of similar movies based on the index
print('Movies suggested for you: \n')
i=1
for movie in sorted_similarity_movies:    
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0] #from index finding the movies
  if (i<30): #run 30 times to give 30 movie
    print(i,'.',title_from_index)#movie is printed with the serial number
    i+=1
```
